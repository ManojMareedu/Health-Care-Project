## Selecting requred columns

```python
var = ['BENE_ID', 'CLM_ID', 'PRVDR_STATE_CD', 'PRVDR_NUM','PRNCPAL_DGNS_CD', 'CLM_TOT_CHRG_AMT', 'CLM_E_POA_IND_SW1' ]

```

## Creating number of claims columns

```python
inpatient['Number_of_Claims'] = inpatient.groupby(['BENE_ID', 'PRNCPAL_DGNS_CD', 'CLM_E_POA_IND_SW1', 'PRVDR_STATE_CD'])['CLM_TOT_CHRG_AMT'].transform('count')
outpatient['Number_of_Claims'] = outpatient.groupby(['BENE_ID', 'PRNCPAL_DGNS_CD', 'PRVDR_STATE_CD'])['CLM_TOT_CHRG_AMT'].transform('count')
```

```python
inpatient['Number_of_Claims'].unique()
```

```python
outpatient['Number_of_Claims'].unique()
```

```python
agg_data_inp = inpatient.groupby(['BENE_ID', 'PRNCPAL_DGNS_CD','CLM_E_POA_IND_SW1','PRVDR_STATE_CD','Number_of_Claims'])['CLM_TOT_CHRG_AMT'].sum().reset_index()
agg_data_out = outpatient.groupby(['BENE_ID', 'PRNCPAL_DGNS_CD','PRVDR_STATE_CD','Number_of_Claims'])['CLM_TOT_CHRG_AMT'].sum().reset_index()

```

# Merging datasets

```python
# Merging agg_data_inp and agg_data_out based on beneficiary_id
merged_data = pd.merge(agg_data_inp, agg_data_out, on='BENE_ID', suffixes=('_inp', '_out'))
```
